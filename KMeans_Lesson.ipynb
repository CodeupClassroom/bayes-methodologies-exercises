```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np

import seaborn as sns
from matplotlib import pyplot as plt

import math
import env

url = f'mysql+pymysql://{env.user}:{env.password}@{env.host}/iris_db'
query = '''
SELECT * FROM measurements JOIN species USING(species_id);
'''

iris = pd.read_sql(query, url)
```

```python
plt.rc('figure', figsize=(16, 9))
plt.rc('font', size=14)
```

```python
X = iris[['petal_length', 'petal_width']]

plt.scatter(X.petal_length, X.petal_width)
plt.gca().set(xlabel='petal length (cm)', ylabel='petal width (cm)', title='Petal Length v Width')
```

```python
starting_points = [(3, 1), (1, 2), (6, 1)]
centers = pd.DataFrame(starting_points, columns=['petal_length', 'petal_width'])
centers
```

```python
X
```

```python
plt.scatter(X.petal_length, X.petal_width)
plt.scatter(centers.petal_length, centers.petal_width, marker='x', s=1000)
```

```python
def distance(p, q):
    return math.sqrt((p.petal_length - q.petal_length)**2 +
                     (p.petal_width - q.petal_width)**2)

def find_cluster(row: pd.Series):
    distances = centers.apply(lambda center: distance(center, row), axis=1)
    return distances.idxmin()
```

```python
X.apply(find_cluster, axis=1)
```

```python
X['cluster'] = X.apply(find_cluster, axis=1)
X
```

```python
for cluster in X.cluster.unique():
    subset = X[X.cluster == cluster]
    plt.scatter(subset.petal_length, subset.petal_width, label=cluster)
plt.legend()
plt.scatter(centers.petal_length, centers.petal_width, marker='x', s=1000, c='black')
```

```python
from sklearn.cluster import KMeans

input_features = ['petal_length', 'petal_width'] 

kmeans = KMeans(n_clusters=8)
kmeans.fit(iris[input_features])
iris['cluster'] = kmeans.predict(iris[input_features])

pd.DataFrame(kmeans.cluster_centers_, columns=input_features)
```
